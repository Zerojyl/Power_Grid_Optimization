```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
config = {
    'data1_path':'./merged_power_data1.csv',
    'data2_path':'./merged_power_data2.csv',
    'data3_path':'./merged_power_data3.csv',
    'data4_path':'./merged_power_data4.csv',
}
```

```python
def float2time(time, dataframe, time_cloumns):
    '''
    time: 参考时间，比如 2019-01-01 00:00:00
    dataframe: 数据表
    time_cloumn: 时间列
    return: 将秒(浮点数)转换为时间
    '''
    ref_time = pd.Timestamp(time)
    for time_cloumn in time_cloumns:
        dataframe[time_cloumn] = ref_time + pd.to_timedelta(dataframe[time_cloumn], unit='s')
    return dataframe
```

```python

def change_data(path):
    data = pd.read_csv(path)
    base_time = '2019-01-01 00:00:00'
    time_cloumns = ['time']
    data = float2time(base_time, data, time_cloumns)
    data.to_csv(path, index=False)
```

```python
change_data(config['data1_path'])
change_data(config['data2_path'])
change_data(config['data3_path'])
change_data(config['data4_path'])
```

```python
def load_show_data(path):
    data = pd.read_csv(path)
    data.set_index('time', inplace=True)
    data.head()
    return data
```
